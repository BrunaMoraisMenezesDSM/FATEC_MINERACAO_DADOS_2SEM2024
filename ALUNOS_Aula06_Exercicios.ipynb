# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
# Importando a biblioteca NumPy
import numpy as np

# Criando do array com 10 números inteiro aleatórios entre 1 e 20
array = np.random.randint(1, 21, size = 10)

# Exibindo array
print("Array original: ", array)

# Extraindo os elementos pares
pares = array[array % 2 == 0]

# Exibindo elementos extraídos
print("Elementos pares: ", pares)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
# Criando dois arrays de 10 números inteiro aleatórios entre 0 e 50
primeiro_array = np.random.randint(1, 51, size = 10)
segundo_array = np.random.randint(1, 51, size = 10)

# Exibindo arrays
print("Primeiro array:", primeiro_array)
print("Segundo array:", segundo_array)

# Soma
soma = primeiro_array + segundo_array
print("Soma:", soma)

# Subtração
subtracao = primeiro_array - segundo_array
print("Subtração:", subtracao)

# Multiplicação
multiplicacao = primeiro_array * segundo_array
print("Multiplicação:", multiplicacao)

# Divisão
divisao = primeiro_array / segundo_array
print("Divisão:", divisao)
```

# 03 - Indexação e Slicing

Dado um array Numpy com 20 números aleatórios entre 10 e 100, extraia os elementos entre a 5ª e a 15ª posição.

```python
# Criando um array com 20 números aleatórios entre 10 e 100
array = np.random.randint(10, 101, size = 20)

# Exibindo o array
print("Array:", array)

# Extraindo os elementos entre a 5ª e a 15ª posição
elementos = array[4:15]

# Exibindo elementos extraídos
print("Elementos da 5ª até a 15ª posição:", elementos)
```

# 04 - Manipulação de Matrizes

Crie uma matriz 3x3 com valores aleatórios entre 0 e 1, e encontre o valor máximo e mínimo.

```python
# Criando uma matriz 3x3 com valores aleatórios entre 0 e 1
matriz = np.random.rand(3, 3)

# Exibindo a matriz
print("Matriz 3x3: \n", matriz)

# Encontrando o valor máximo e mínimo
valor_maximo = np.max(matriz)
valor_minimo = np.min(matriz)

# Exibindo os valores máximo e mínimo
print("Valor máximo:", valor_maximo)
print("Valor mínimo:", valor_minimo)
```

# 05 - Transformações em Arrays

# 07 - Criação de DataFrame com Dados Sintéticos

Crie um DataFrame fictício de vendas contendo colunas como Produto, Quantidade (entre 1 e 20), e Preço (entre 10 e 100).

```python
# Criando um DataFrame fictício de vendas
df_vendas = pd.DataFrame({'Produto': [ 'Perfume', 'Desodorante', 'Creme Corporal', 'Sabonete', 'Pasta de dente', 'Shampoo', 'Condicionador', 'Batom', 'Loção Hidratante',
                          'Esfoliante', 'Creme Facial', 'Protetor Solar', 'Esmalte', 'Toalha', 'Escova de Dentes'],
                           'Quantidade': np.random.randint(1, 21, size = 15),
                           'Preço': np.round(np.random.uniform(10, 101, size = 15), 2)})

# Exibindo o DataFrame
df_vendas
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
# Exibindo vendas onde a quantidade é maior que 10
df_vendas[df_vendas["Quantidade"] > 10]
```

# 09 - Operações em Colunas

Adicione uma coluna ao DataFrame com o valor total da venda (Quantidade * Preço).

```python
# Criando a coluna Valor Total
valor_total = df_vendas['Quantidade'] * df_vendas['Preço']

# Adicionando a coluna ao Data Frame
df_vendas = pd.concat([df_vendas, valor_total.rename("Valor Total")], axis = 1)

# Exibindo o DataFrame
df_vendas
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
# Agrupando os dados pelo campo Produto e somando as quantidades vendidas
df_vendas.groupby("Produto")['Quantidade'].sum()
```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
# Importando a biblioteca matplotlib e o módulo pyplot
from matplotlib import pyplot as plt

# Criando arrays para meses e preços
meses = np.arange(1, 13)
precos = np.random.uniform(10, 101, size = 12).round(2)

# Plotando gráfico de linha
plt.plot(meses, precos, marker = 'o')

# Personalizando o gráfico
plt.title('Variação de preços de um produto ao longo de 12 Meses')
plt.xlabel('Meses')
plt.ylabel('Preço')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

# 14 - Gráfico de Dispersão

Crie um gráfico de dispersão que mostre a relação entre altura e peso em um conjunto de dados fictício.

```python
# Gerando arrays de 100 pessoas com alturas (em cm) e pesos (em kg)
alturas = np.random.uniform(110, 200, size = 100)
pesos = np.random.uniform(40, 100, size = 100)

# Criando figura para o gráfico
plt.figure(figsize = (9, 4))

# Plotando o gráfico de dispersão
grafico_dispersao = plt.scatter(alturas, pesos)

# Personalizando o gráfico
plt.title('Relação entre altura e peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

# 16 - Gráfico de Barras com Seaborn

Use Seaborn para criar um gráfico de barras com os mesmos dados fictícios de quantidade vendida por produto.

```python
# Importando a biblioteca Seaborn
import seaborn as sns

# Criando um DataFrame com os dados dos exercícios anteriores
df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades
})

# Criando figura para o gráfico
plt.figure(figsize = (12, 6))

# Plotando gráfico de barra com Seaborn
grafico_barra = sns.barplot(x = 'Produto', y = 'Quantidade', data = df_vendas, palette = 'husl')

# Personalizando o gráfico
plt.title('Quantidade vendida por produto')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.xticks(rotation = 45, ha = 'right')

# Exibindo o gráfico
plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
# Criando figura para o gráfico
plt.figure(figsize = (10, 6))

# Criando um DataFrame com os dados dos exercícios anteriores
df_dados = pd.DataFrame({
    'Altura': alturas,
    'Peso': pesos
})

# Plotando o gráfico de dispersão com linha de regressão
grafico_dispersao = sns.regplot(x = 'Altura', y = 'Peso', data = df_dados, scatter_kws = {'alpha': 0.7}, line_kws = {'color': 'orange'})

# Personalizando o gráfico
plt.title('Relação entre altura e peso com linha de regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
# Gerando um DataFrame fictício com variáveis aleatórias
df_dados = pd.DataFrame({
    'primeiro': np.random.rand(100),
    'segundo': np.random.rand(100),
    'terceiro': np.random.rand(100),
    'quarto': np.random.rand(100),
    'quinto': np.random.rand(100)
})

# Calculando a matriz de correlação
matriz_correlacao = df_dados.corr()

# Criando figura para o gráfico
plt.figure(figsize = (10, 8))

# Plotando o heatmap
sns.heatmap(matriz_correlacao, annot = True, cmap = 'plasma', fmt = '.2f', linewidths = 0.5, linecolor = 'black')

# Personalizando o gráfico
plt.title('Heatmap com a correlação entre variáveis')

# Exibindo o gráfico
plt.show()
```
